2. Identifica el país con la mayor densidad de población y el país con la menor población en el conjunto de datos.

```python
df.head(2)
```

```python
# Encuentra el país con la mayor densidad de población
pais_max_densidad = df.loc[df['density'].idxmax(), 'country']
max_densidad = df['density'].max()


print(f"País con mayor densidad de población: {pais_max_densidad} (Densidad: {max_densidad} personas por km²)")
```

```python
# Encuentra el país con la mayor densidad de población
pais_min_densidad = df.loc[df['density'].idxmin(), 'country']
min_densidad = df['density'].min()


print(f"País con menor densidad de población: {pais_min_densidad} (Densidad: {min_densidad} personas por km²)")
```
