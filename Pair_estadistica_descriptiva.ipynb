4. Realiza un análisis de la mortalidad infantil (infant_mortality) y clasifica los países en tres categorías: baja, moderada y alta mortalidad infantil. Belen

```python
# Definir los límites para las categorías
baja = 10  
moderada = 20
alta =30
df.loc[df['infantmortality'] < baja, 'mortalidad_infantil_categoria'] = 'Baja'
if moderada is not None:
    df.loc[df['infantmortality'] > moderada, 'mortalidad_infantil_categoria'] = 'Moderada'
if alta is not None:
    df.loc[df['infantmortality'] > alta, 'mortalidad_infantil_categoria'] = 'Alta'
df.head(2)
```

```python
# Mostrar los resultados
print(df[['country', 'infantmortality', 'mortalidad_infantil_categoria']])
```

```python

conteo_categorias = df['mortalidad_infantil_categoria'].value_counts()
plt.bar(conteo_categorias.index, conteo_categorias)

plt.xlabel('Categoría de Mortalidad Infantil')
plt.ylabel('Número de Países')
plt.title('Distribución de Mortalidad Infantil por Categoría')
plt.show()
```

9. Genera un mapa de calor (heatmap) de la matriz de correlación de todas las variables numéricas y discute las relaciones más destacadas.

```python
matriz_corr = df_numericas.corr()

# creamos una matriz booleana de máscara
mask = np.triu(np.ones_like(matriz_corr, dtype=bool))

#heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(matriz_corr, annot=True, cmap='coolwarm', fmt=".1f", vmax = 1, vmin = -1, mask= mask);
plt.title('Mapa de Calor de Matriz de Correlación')

```
